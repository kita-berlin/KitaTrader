```python
from Api.CoFu import *
from Api.KitaApiEnums import RunMode
from Robots.Martingale import Martingale
from datetime import datetime
```

```python
# 1. Set the parameters of the platform
# region
robot = Martingale()  # Define here which robot should be used
robot.robot = robot  # type:ignore
```

```python

# robot.AllDataStartUtc = datetime.min means earliest possible what the source can provide
# End datetime always is yesterday. Cannot be today because today's data are not complete yet
robot.AllDataStartUtc = datetime.strptime("3.1.2006", "%d.%m.%Y")

# Platform mode how the robot will be used by the platform
# Other possibilities are (not yet implemented): RealTime (live trading),
# VisualBacktesting, BruteForceOptimization, GeneticOptimization, WalkForwardOptimization
robot.RunningMode = RunMode.SilentBacktesting

# The path where the robot can store data
robot.DataPath = "..\\QuantConnect\\MyLean\\MyData\\cfd"

# Paper trading initial account settings
robot.AccountInitialBalance = 10000
robot.AccountLeverage = 500
# until full currency conversion is implemented, the quote currency of the traded symbol is used as the account currency
robot.AccountCurrency = "EUR"
# endregion
```

```python
# 2. # Define the backtest time window
# region
robot.BacktestStartUtc = datetime.strptime("1.1.2024", "%d.%m.%Y")
#robot.BacktestEndUtc = datetime.strptime("10.12.2024", "%d.%m.%Y")
robot.BacktestEndUtc = datetime.max
# endregion
```

```python
# 3. Initialize the platform and the robot
# region
robot.init()  # type: ignore
# endregion
```

```python
# 4. Start the platform and the robot
# region
robot.start()  # type: ignore
# endregion
```

```python
# 5. loop over the give time range
# region
while True:
    if robot.tick():
        break
# endregion
```

```python
# 6. Stop the robot and the platform
# region
robot.stop()
# endregion
```
